```python
df = pd.read_csv('hiring.csv')
df
```

```python
import seaborn as sns
sns.pairplot(df)
```

```python
df.isnull().sum()
```

```python
import numpy as np

word_to_num = {
    'zero': 0, 'one': 1, 'two': 2, 'three': 3,
    'four': 4, 'five': 5, 'six': 6, 'seven': 7,
    'eight': 8, 'nine': 9, 'ten': 10, 'eleven': 11
}

df['experience'] = df['experience'].replace(word_to_num)
```

```python
df['experience'] = df['experience'].fillna(method='bfill')
```

```python
df
```

```python
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(df['test_score(out of 10)'].mean())
```

```python
df.isnull().sum()
```

```python
X = df.drop('salary($)', axis=1)
y = df['salary($)']
```

```python
reg = linear_model.LinearRegression()
reg.fit(X,y)
```

```python
reg.predict([[12, 9.2, 8.9]])
```
